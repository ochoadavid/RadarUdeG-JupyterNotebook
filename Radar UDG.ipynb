```python
import folium
from folium import plugins
from scipy.ndimage import imread
import urllib.request
```

```python

urllib.request.urlretrieve('http://astro.iam.udg.mx/radar/imagen_actuali.png', 'imagen_actuali.png');
urllib.request.urlretrieve('http://astro.iam.udg.mx/radar/imagen_actualv.png', 'imagen_actualv.png');
```

```python
# boundary of the image on the map
min_lon = -105.7850428061314
max_lon = -100.9866246743374
min_lat = 18.4171756291133
max_lat = 22.93334615140791

# create the map
map_ = folium.Map(location=[20.66, -103.36],
                  tiles='Stamen Terrain', zoom_start = 11)

# read in png file to numpy array
img_intensidad = imread('./imagen_actuali.png')
img_velocidad = imread('./imagen_actualv.png')

Intensidad = plugins.ImageOverlay(img_intensidad, opacity=0.8, 
        bounds =[[min_lat, min_lon], [max_lat, max_lon]],
        attr = 'Intensidad')
Velocidad = plugins.ImageOverlay(img_velocidad, opacity=0.8, 
        bounds =[[min_lat, min_lon], [max_lat, max_lon]],
        attr = 'Velocidad')

# add a name to the layer
Intensidad.layer_name = 'Intensidad'
Velocidad.layer_name = 'Velocidad'

# Overlay the image
map_.add_child(Velocidad)
map_.add_child(Intensidad)

# add the layer control
folium.LayerControl().add_to(map_)

map_
```
